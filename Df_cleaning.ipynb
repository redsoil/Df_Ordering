```python
import pandas as pd
```

```python
df = pd.read_csv('df.csv')
```

```python
df
```

```python
def dollar_remover(a):

  a = a.replace('$', '')
  return a
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(dollar_remover)
```

```python
df
```

```python
def k_remover(b):

  b = b.replace('K', '')
  return b
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(k_remover)
```

```python
df
```

```python
df = df[df['Salary Estimate'] != '-1']
```
